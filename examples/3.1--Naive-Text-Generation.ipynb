```python
import random

import pandas as pd
import seaborn as sns

import cltrier_nlp as nlp
```

```python
# vocab

PROPN = lambda: random.choice(["mary", "peter"])
N = lambda: random.choice(["computer", "table", "chair"])
DET = lambda: random.choice(["a", "the"])

iV = lambda: random.choice(["sits", "stands"])
V = lambda: random.choice(["loves", "hates"])
P = lambda: random.choice(["on", "under"])
```

```python
# grammar

NP = lambda: random.choice([[DET(), N()], PROPN()])
PP = lambda: [P(), NP()]
VP = lambda: random.choice([[iV(), PP()], [V(), NP()]])

S = lambda: [NP(), VP()] 
```

```python
for _ in range(20):
    print(S())
```
